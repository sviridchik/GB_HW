```python
import random
import pandas as pd
import numpy as np
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
pd.get_dummies(data['whoAmI'])
```

```python
titles = data.whoAmI.unique ()
titles
```

```python
titles[0]
```

```python
val = []
for i in titles:
    val.append(list(np.where(data['whoAmI']==i, 1, 0)))
val
```

```python
res = dict(zip(titles,val))
res
```

```python
pd.DataFrame(res)
```

```python
answer = data.join(pd.DataFrame(res)) 
answer
```
