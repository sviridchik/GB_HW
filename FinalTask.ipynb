### Условие 3: 3 задача
+ Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
+ Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df[['price', 'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'zipcode',
       'lat', 'long', 'sqft_living15', 'sqft_lot15']].corr()
corr_matrix = np.round(corr_matrix,1)
corr_matrix[np.abs(corr_matrix<0.3)]=0
corr_matrix
```

```python
sns.heatmap(corr_matrix, annot=True, linewidth = 0.5, cmap='coolwarm')
```

## Выводы 
+ Согласно матрице корреляция можно сделать вывод, что на стоимость сильно влияют  квадратные метры (размер)и grade
+ Чуть меньше влияют sqft_living15 и sqft_above

```python

plt.scatter(df['sqft_living'], df['price'])
plt.xlabel('sqft_living')
plt.ylabel('Стоимость')
plt.title('Распределение цен в зависимости от метража');
```

## Выводы
+ Ожидаемая связь: при увеличении метража повышается стоимость. Исключение выброс на самой большой квартире. Возможно из-за размеров было сложно найти покупателя , ведь ее содержание будет не простым и из-за это снизили стоимость.А возможно просто выброс
+ Большинство домов имеют метраж до 8000
